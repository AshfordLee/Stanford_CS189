```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from scipy.stats import norm
from mpl_toolkits.mplot3d import Axes3D
```

# Plot 1-D normal pdf

```python
mu=1
sigma=1
x=np.linspace(-4,4,100)

pdf=norm.pdf(x,mu,sigma)

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, 'b-', linewidth=2, label=f'μ={mu}, σ={sigma}')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('1-d normal pdf')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

# Plot 2-D multivariable pdf

```python
mean=[1,1]

cov=[[1,0],[0,2]]

x=np.linspace(-3,3,100)
y=np.linspace(-3,3,100)
X,Y=np.meshgrid(x,y)
pos=np.dstack((X,Y))

rv=multivariate_normal(mean,cov)
Z=rv.pdf(pos)

fig=plt.figure(figsize=(24,10))

ax1=fig.add_subplot(121,projection='3d')
ax1.plot_surface(X,Y,Z,cmap='viridis',alpha=0.8)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('pdf')
ax1.set_title('3D_surface')

ax2=fig.add_subplot(122)
contour=ax2.contour(X,Y,Z)
ax2.clabel(contour, inline=True, fontsize=8)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Isocontours')
ax2.grid(True)



plt.tight_layout()
plt.show()
```

# difference of two PDFS

```python
mean1=[0,2]

cov1=[[2,1],[1,1]]

mean2=[2,0]

cov2=[[2,1],[1,4]]


rv1=multivariate_normal(mean1,cov1)
rv2=multivariate_normal(mean2,cov2)

x=np.linspace(-3,3,100)
y=np.linspace(-3,3,100)
X,Y=np.meshgrid(x,y)
pos=np.dstack((X,Y))

combined_rv=rv1.pdf(pos)-rv2.pdf(pos)


fig=plt.figure(figsize=(24,10))

ax1=fig.add_subplot(121,projection='3d')
ax1.plot_surface(X,Y,combined_rv,cmap='viridis',alpha=0.8)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('rv1-rv2')
ax1.set_title('3D_surface')

ax2=fig.add_subplot(122)
contour=ax2.contour(X,Y,combined_rv)
ax2.clabel(contour, inline=True, fontsize=8)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Isocontours')
ax2.grid(True)



plt.tight_layout()
plt.show()
```
