```python
import matplotlib.pyplot as plt
import pandas as pd

# 创建 2x2 的子图布局
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 选择要绘制的定量变量
variables = ['Apps', 'Accept', 'Enroll', 'Outstate']

# 为每个变量绘制直方图，使用不同的 bins 数量
bins_list = [10, 20, 15, 25]  # 为每个变量设置不同的 bins 数量

for i, (var, bins) in enumerate(zip(variables, bins_list)):
    row = i // 2  # 行索引
    col = i % 2   # 列索引
    
    # 在对应的子图位置绘制直方图
    college[var].plot.hist(bins=bins, ax=axes[row, col])
    axes[row, col].set_title(f'{var} (bins={bins})')
    axes[row, col].set_xlabel(var)
    axes[row, col].set_ylabel('Frequency')

plt.tight_layout()  # 自动调整子图间距
plt.show()
```
